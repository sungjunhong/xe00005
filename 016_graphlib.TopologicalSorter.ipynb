### 016 수강할 과목의 순서를 구하려면?

```python
from graphlib import TopologicalSorter  # python>=3.9

ts = TopologicalSorter()

# 규칙1
ts.add('영어 중급', '영어 초급')
ts.add('영어 고급', '영어 중급')

# 규칙2
ts.add('영어 문법', '영어 중급')
ts.add('영어 고급', '영어 문법')

#규칙3
ts.add('영어 회화', '영어 문법')

print(list(ts.static_order()))
```

```python
from graphlib import TopologicalSorter  # python>=3.9

ts = TopologicalSorter()

ts.add('점화', '냄비에 물 붓기')

ts.add('라면 넣기', '점화')
ts.add('라면 넣기', '라면 봉지 뜯기')

ts.add('수프 넣기', '점화')
ts.add('수프 넣기', '라면 봉지 뜯기')

ts.add('계란 풀어넣기', '점화')
ts.add('계란 풀어넣기', '라면 넣기')
ts.add('계란 풀어넣기', '수프 넣기')

print(list(ts.static_order()))
```
