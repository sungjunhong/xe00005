### 014 점수에 따른 학점을 구하려면?

```python
result = []

for score in [33, 99, 77, 70, 89, 90, 100]:
    if score >= 90:
        result.append('A')
    elif score >= 80:
        result.append('B')
    elif score >= 70:
        result.append('C')
    elif score >= 60:
        result.append('D')
    else:
        result.append('F')

print(result)
```

```python
import bisect

result = []

for score in [33, 99, 77, 70, 89, 90, 100]:
    pos = bisect.bisect([60, 70, 80, 90], score)    # bisect.bisect_right()
    grade = 'FDCBA'[pos]
    result.append(grade)

print(result)
```

```python
import bisect

result = []

for score in [33, 99, 77, 70, 89, 90, 100]:
    pos = bisect.bisect_left([60, 70, 80, 90], score)
    grade = 'FDCBA'[pos]
    result.append(grade)

print(result)
```
